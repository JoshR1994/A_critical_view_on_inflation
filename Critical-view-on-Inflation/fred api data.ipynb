Building a new df and dataset of several interesting quantities:

```python
df_unrate=pd.DataFrame(fred.get_series(series_id='UNRATE'))
df_unrate.columns=['Unemployment Rate USA in %']
```

```python
df_wil=pd.DataFrame(fred.get_series(series_id='WILL5000INDFC'))
df_wil.columns=['Daily Close of Wilshire 5000']
```

```python
df_nasdaq=pd.DataFrame(fred.get_series(series_id='NASDAQCOM'))
df_nasdaq.columns=['Daily Close NASDAQ']
```

```python
df_nikkei=pd.DataFrame(fred.get_series(series_id='NIKKEI225'))
df_nikkei.columns=['Daily Close Nikkei']
```

```python
df_ms=pd.DataFrame(fred.get_series(series_id='M2SL'))
df_ms.columns=['Money Supply USA in Bil. of $']
```

```python
df_sp=pd.DataFrame(fred.get_series(series_id='SP500'))
df_sp.columns=['Daily Close S&P500 (2000=100)']
# df_sp=df_sp[df_sp['S&P500 Index Value (2000=100)'].isna()==False]
```

```python
df_capgdprate=pd.DataFrame(fred.get_series(series_id='DDDM01USA156NWDB'))
df_capgdprate.columns=['USA Stock Market Capitalization to GDP ratio in %']
```

```python
df_cpi=pd.DataFrame(fred.get_series(series_id='CPIAUCSL'))
df_cpi.columns=['CPI Index (1982-1984=100)']
```

```python
df_test=pd.merge(df_wil,df_nasdaq,left_index=True, right_index=True,how='outer')
```

```python
df_test=pd.merge(df_test,df_nikkei,left_index=True, right_index=True,how='outer')
```

```python
df_stock_data=pd.merge(df_test,df_sp,left_index=True, right_index=True,how='outer')
```

```python
df_test1=pd.merge(df_unrate,df_ms,left_index=True, right_index=True,how='outer')
```

```python
df_test1=pd.merge(df_test1,df_capgdprate,left_index=True, right_index=True,how='outer')
```

```python
df_fed_data=pd.merge(df_test1,df_cpi,left_index=True, right_index=True,how='outer')
```

Plotting the data:

```python
df_final=pd.merge(df_stock_data,df_fed_data,left_index=True, right_index=True,how='outer')
```

```python
df_final.to_csv('fred_data.csv')
```

```python
for columns in df_final.columns:
    df_final[[columns]].dropna().plot()
```

Correlation matrix:

```python
round(df_final.corr(),2)
```

Scatter plots showing correlation of different entries in dataframe:

```python
df=df_final[['Money Supply USA in Bil. of $','Daily Close S&P500 (2000=100)']]
df=df.dropna()

plt.title('Money Supply & S&P500 Index, correlation factor 0.97')
plt.xlabel('Money Supply USA in Bil. of $')
plt.ylabel('S&P500 Index Value')
x=df['Money Supply USA in Bil. of $']
y=df['Daily Close S&P500 (2000=100)']
plt.scatter(x,y,s=5)
a, b = np.polyfit(x,y,1)
plt.plot(x,a*x+b)
```

```python
df=df_final[['Money Supply USA in Bil. of $','Daily Close of Wilshire 5000']]
df=df.dropna()

plt.title('Money Supply & Wilshire 5000, correlation factor 0.98')
plt.xlabel('Money Supply USA in Bil. of $')
plt.ylabel('Daily Close of Wilshire 5000')
x=df['Money Supply USA in Bil. of $']
y=df['Daily Close of Wilshire 5000']
plt.scatter(x,y,s=5)
a, b = np.polyfit(x,y,1)
plt.plot(x,a*x+b)
```

```python
df=df_final[['Money Supply USA in Bil. of $','USA Stock Market Capitalization to GDP ratio in %']]
df=df.dropna()

plt.title('Money Supply & Market Cap to GDP ratio in the USA, correlation factor 0.8')
plt.xlabel('Money Supply USA in Bil. of $')
plt.ylabel('Daily Close of Wilshire 5000')
x=df['Money Supply USA in Bil. of $']
y=df['USA Stock Market Capitalization to GDP ratio in %']
plt.scatter(x,y,s=5)
a, b = np.polyfit(x,y,1)
plt.plot(x,a*x+b)
```

```python
df=df_final[['Money Supply USA in Bil. of $','CPI Index (1982-1984=100)']]
df=df.dropna()

plt.title('Money Supply & CPI in the USA, correlation factor 0.91')
plt.xlabel('Money Supply USA in Bil. $')
plt.ylabel('CPI Index (1982-1984=100)')
x=df['Money Supply USA in Bil. of $']
y=df['CPI Index (1982-1984=100)']
plt.scatter(x,y,s=5)
a, b = np.polyfit(x,y,1)
plt.plot(x,a*x+b)
```
